```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("ML case Study.csv", header=0)
```

```python
df.head()
```

```python
college = pd.read_csv("Colleges.csv", header=0)
```

```python
college.head()
```

```python
Tier1 = college['Tier 1'].tolist()
Tier2 = college['Tier 2'].tolist()
Tier3 = college['Tier 3'].tolist()
```

```python
Tier3
```

```python
for item in df.College:
    if item in Tier1:
        df['College'].replace(item,1,inplace=True)
    elif item in Tier2:
        df['College'].replace(item,2,inplace=True)
    elif item in Tier3:
        df['College'].replace(item,3,inplace=True)
```

```python
df.head()
```

```python
city = pd.read_csv("cities.csv", header=0)
```

```python
city.head()
```

```python
metro_city = city['Metrio City'].tolist()
non_metro_city = city['non-metro cities'].tolist()
```

```python
metro_city
```

```python
non_metro_city
```

```python
for item in df.City:
    if item in non_metro_city:
        df['City'].replace(item,"non metro city",inplace=True)
    elif item in metro_city:
        df['City'].replace(item,"metro city",inplace=True)
```

```python
df = pd.get_dummies(df, columns = ['City','Role'])
```
